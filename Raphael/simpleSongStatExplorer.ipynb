```python
f = open('../mxm_dataset_test.txt/mxm_dataset_test.txt', 'r')
data = f.readlines()
f.close
data = [line for line in data if line[0] != "#"]
words = [word for word in data[0].split(",")]

def songStat(song):
    wordDict = {}
    song = song.split(",")
    wordDict['track_id'] = song[0]
    wordDict['mxm_track_id'] = song[1]
    for wordStat in song[2:]:
        wordID = int(wordStat.split(":")[0])
        wordCnt = int(wordStat.split(":")[1])
        wordDict[words[wordID]] = wordCnt
    return wordDict
```
