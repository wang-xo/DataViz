```python
import pandas as pd
import math
df = pd.read_csv("../merged_data/words_songs_matched_90s.csv", sep=",")
```

```python
df = df.drop(columns=['song_id', 'artist_id'])
```

```python
df[['word','count']].groupby('word').agg(sum).reset_index().sort_values(['count'],ascending=False)
```

```python
df.groupby('word')['count'].transform(sum).max()
```

```python
df['max_count_in_data'] = df.groupby('word')['count'].transform(sum).max()
df['word_count_in_data'] = df.groupby('word')['count'].transform(sum)
df['tf'] = df['word_count_in_data']/df['max_count_in_data']
print('tf domain:', [df['tf'].min(),df['tf'].max()])
```

```python
df
```

```python
df['df'] = df.groupby(['word'])['word'].transform('count')
track_count = len(df['track_id'].unique())
print(track_count)
def calIDF(df, track_count):
    return math.log2(track_count/df)
df['idf'] = df['df'].apply(calIDF, track_count=track_count)
print('df domain:', [df['df'].max(),df['df'].min()])
print('idf domain:', [df['idf'].min(),df['idf'].max()])
```

```python
df['tfidf'] = df['tf']*df['idf']
print('tfidf domain:', [df['tfidf'].min(),df['tfidf'].max()])
```
