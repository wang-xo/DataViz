```python
import pandas as pd
```

```python
df = pd.read_csv("words_songs_matched.csv", comment='#', header=0, sep=',')
```

```python
#Create list of words
wordlist = df.word.unique()
```

```python
df.head()
```

```python
#Create empty matrix
matrix = pd.DataFrame(data= 0, columns= wordlist, index = wordlist)
matrix
```

```python
#Creating subset
tinydf = df[(df.artist_name == 'Metallica') & (df.year == 1991)]
```

```python
tinydf2 = tinydf[['word', 'count', 'title', 'artist_name', 'year']].reset_index(drop=True)
```

```python
#Creating songgroups
dfgroup = tinydf2.groupby(['artist_name', 'title'])['word'].unique().reset_index()
print(dfgroup)
```

```python
#Creating a list from the dataframe
grouplist = dfgroup.to_numpy().tolist()
grouplist
```

```python
#Populating matrix
matrix = pd.DataFrame(data= 0, columns= wordlist, index = wordlist)

for word in wordlist:
    for index1, row1 in enumerate(grouplist):
        if word in row1[2]:
            for songword in row1[2]:
                matrix.loc[word, songword] += 1
```

```python
matrix
```

```python
matrix.to_csv('metal_martix.csv')
```
