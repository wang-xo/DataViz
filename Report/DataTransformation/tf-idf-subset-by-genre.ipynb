```python
import pandas as pd
import math
import numpy as np
df = pd.read_csv("../merged_data/words_songs_matched_90s.csv", sep=",")
g = pd.read_csv("../Songs_genre/msd_tagtraum_cd2c.cls.csv", sep="\t")
df = df.merge(g, left_on="track_id", right_on="trackId")
df = df.drop(columns=['minority-genre','trackId'])
```

TFIDF Style Metric

```python
df.head()
```

```python
#max frequency by year
for year in df.year.unique():
    print(year)
    df.loc[df['year'] == year, 'max_f_by_year'] = df.loc[df['year'] == year].groupby('word')['count'].transform(sum).max()
df.head()
```

```python
#max frequency by genre
for g in df['majority-genre'].unique():
    print(g)
    df.loc[df['majority-genre'] == g, 'max_f_by_genre'] = df.loc[df['majority-genre'] == g].groupby('word')['count'].transform(sum).max()
df.head()
```

```python
#frequency by year
df['f_by_year'] = df.groupby(['word','year'])['count'].transform(sum)
df.head()
```

```python
#frequency by genre
df['f_by_genre'] = df.groupby(['word','majority-genre'])['count'].transform(sum)
df.head()
```

```python
#tf by year
df['tf_by_year'] = df['f_by_year']/df['max_f_by_year']
print('tf domain:', [df['tf_by_year'].min(),df['tf_by_year'].max()])
```

```python
#tf by genre
df['tf_by_genre'] = df['f_by_genre']/df['max_f_by_genre']
print('tf domain:', [df['tf_by_genre'].min(),df['tf_by_genre'].max()])
```

```python
#track_count by year
for year in df.year.unique():
    print(year)
    df.loc[df['year'] == year, 'track_count_by_year'] = df.loc[df['year'] == year]['track_id'].nunique()
df.head()
```

```python
#track_count by genre
for g in df['majority-genre'].unique():
    print(g)
    df.loc[df['majority-genre'] == g, 'track_count_by_genre'] = df.loc[df['majority-genre'] == g]['track_id'].nunique()
df.head()
```

```python
#df by year
df['df_by_year'] = df.groupby(['word','year'])['word'].transform('count')
df.head()
```

```python
#df by genre
df['df_by_genre'] = df.groupby(['word','majority-genre'])['word'].transform('count')
df.head()
```

```python
#idf
df['idf_by_year'] = np.log2(df['track_count_by_year']/df['df_by_year'])
df['idf_by_genre'] = np.log2(df['track_count_by_genre']/df['df_by_genre'])
```

```python
df['tfidf_by_year'] = df['tf_by_year']*df['idf_by_year']
print('tfidf domain:', [df['tfidf_by_year'].min(),df['tfidf_by_year'].max()])
df['tfidf_by_genre'] = df['tf_by_genre']*df['idf_by_genre']
print('tfidf domain:', [df['tfidf_by_genre'].min(),df['tfidf_by_genre'].max()])
```
